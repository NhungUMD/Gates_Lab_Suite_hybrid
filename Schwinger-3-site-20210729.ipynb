```python
def generate_SCHW_test_circuit(ionorder):
    L=[str(i) for i in ionorder];
    nqubit=6;
    statei=Quantum_State(nqubit)
    Set_Mapping(ionorder)
    mu = 0.1;
    x = 0.6;
    dt = 1;
    N = 10;

    file_name="3-site_010101_dt="+str(dt)+"_eo"+''.join(L)+"_2.txt"
    write_file=open(circuit_path+file_name,'w')
    cal_name="3-site-cal"+''.join(L)+".txt"
    write_cal=open(circuit_path+cal_name,'w')
    cal_list=[]

    C1=Quantum_Circuit(nqubit,"3-spatial")

    C1.Add_Gate(Quantum_Gate("SK1X",1,angle=1))
    C1.Add_Gate(Quantum_Gate("SK1X",3,angle=1))
    C1.Add_Gate(Quantum_Gate("SK1X",5,angle=1))
    ## odd-leading terms
    for i in range(0,nqubit-1,2):
    #     print(i)
        C1.Add_Gate(Quantum_Gate("XA",i,i+1,angle=0.5*x*dt/pi))  
        C1.Add_Gate(Quantum_Gate("RZ",i,angle=2))
        C1.Add_Gate(Quantum_Gate("RZ",i+1,angle=2))
        C1.Add_Gate(Quantum_Gate("XA",i,i+1,angle=0.5*x*dt/pi))
        C1.Add_Gate(Quantum_Gate("RZ",i,angle=-2))
        C1.Add_Gate(Quantum_Gate("RZ",i+1,angle=-2))
        seq = C1.GatesLab_Sequence()
        write_file.write(str(seq)+"\n")
        M1=Quantum_Circuit(nqubit,"3-spatial")
        M1.Add_Gate(Quantum_Gate("XX",i,i+1))
        cal_seq = M1.GatesLab_Sequence()
        cal_seq_str = str(cal_seq)
        if (cal_seq_str[2]>cal_seq_str[3]):
            cal_seq_str = cal_seq_str[0:2]+cal_seq_str[3]+cal_seq_str[2]+cal_seq_str[4:]
        if cal_seq_str not in cal_list:
            cal_list.append(cal_seq_str)
            write_cal.write(cal_seq_str+"\n")

    #
    for i in range(1,nqubit-1,2):
    #     print(i)
        C1.Add_Gate(Quantum_Gate("XA",i,i+1,angle=0.5*x*dt/pi))  
        C1.Add_Gate(Quantum_Gate("RZ",i,angle=2))
        C1.Add_Gate(Quantum_Gate("RZ",i+1,angle=2))
        C1.Add_Gate(Quantum_Gate("XA",i,i+1,angle=0.5*x*dt/pi))
        C1.Add_Gate(Quantum_Gate("RZ",i,angle=-2))
        C1.Add_Gate(Quantum_Gate("RZ",i+1,angle=-2))
        seq = C1.GatesLab_Sequence()
        write_file.write(str(seq)+"\n")
        M1=Quantum_Circuit(nqubit,"3-spatial")
        M1.Add_Gate(Quantum_Gate("XX",i,i+1))
        cal_seq = M1.GatesLab_Sequence()
        cal_seq_str = str(cal_seq)
        if (cal_seq_str[2]>cal_seq_str[3]):
            cal_seq_str = cal_seq_str[0:2]+cal_seq_str[3]+cal_seq_str[2]+cal_seq_str[4:]
        if cal_seq_str not in cal_list:
            cal_list.append(cal_seq_str)
            write_cal.write(cal_seq_str+"\n")
    ## ZZ and Z terms
    for i in range(nqubit):
        C1.Add_Gate(Quantum_Gate("SK1Y",i,angle=-2))
    for i in range(1,nqubit-1):
        for j in range(i):
    #         print(j,i)
    #         print(nqubit-i-1)
            C1.Add_Gate(Quantum_Gate("XA",i,j,angle=0.5*(nqubit-i-1)*dt/pi))
            M1=Quantum_Circuit(nqubit,"3-spatial")
            M1.Add_Gate(Quantum_Gate("XX",i,j))
            cal_seq = M1.GatesLab_Sequence()
            cal_seq_str = str(cal_seq)
            if (cal_seq_str[2]>cal_seq_str[3]):
                cal_seq_str = cal_seq_str[0:2]+cal_seq_str[3]+cal_seq_str[2]+cal_seq_str[4:]
            if cal_seq_str not in cal_list:
                cal_list.append(cal_seq_str)
                write_cal.write(cal_seq_str+"\n")
        seq = C1.GatesLab_Sequence()
        write_file.write(str(seq)+"\n")
    for i in range(nqubit):
        C1.Add_Gate(Quantum_Gate("SK1Y",i,angle=2))

    C1.Add_Gate(Quantum_Gate("AZ",0,angle=-(mu+3)*dt/pi)) 
    C1.Add_Gate(Quantum_Gate("AZ",1,angle=(mu-2)*dt/pi))
    C1.Add_Gate(Quantum_Gate("AZ",2,angle=-(mu+2)*dt/pi)) 
    C1.Add_Gate(Quantum_Gate("AZ",3,angle=(mu-1)*dt/pi))
    C1.Add_Gate(Quantum_Gate("AZ",4,angle=-(mu+1)*dt/pi))
    C1.Add_Gate(Quantum_Gate("AZ",5,angle=mu*dt/pi))
    seq = C1.GatesLab_Sequence()
    write_file.write(str(seq))
    write_file.close()
    write_cal.close()
    return
```

```python
def generate_SCHW3_block_circuit(ionorder,dt=1):
    L=[str(i) for i in ionorder];
    nqubit=6;
    statei=Quantum_State(nqubit)
    Set_Mapping(ionorder)
    Nstep=7

    file_name="SCHW3_010101_dt="+str(dt)+"_eo"+''.join(L)+".txt"
    write_file=open(circuit_path+file_name,'w')
    sim_file_name="donotrun_010101_dt="+str(dt)+"_eo"+''.join(L)+".txt"
    write_sim_file=open(circuit_path+sim_file_name,'w')
    
    B1=Quantum_Circuit(nqubit,"SCHW3block")
    B1.Add_Gate(Quantum_Gate("SK1X",1,angle=1))
    B1.Add_Gate(Quantum_Gate("SK1X",3,angle=1))
    B1.Add_Gate(Quantum_Gate("SK1X",5,angle=1))
    seq = B1.GatesLab_Sequence()
    seq_str=str(seq)
    seq_sim_str=seq_str
    write_file.write(seq_str+"\n")
    write_sim_file.write(seq_sim_str+"\n")
    Udt_str=":SCHW3"+''.join(L)+str(dt)
    Udt_sim_str=generate_Udt_SCHW_circuit(ionorder,True)
    for i in range(Nstep):
        seq_str+=Udt_str
        seq_sim_str=seq_sim_str+":"+Udt_sim_str
        write_file.write(seq_str+"\n")
        write_sim_file.write(seq_sim_str+"\n")
    write_file.close()
    write_sim_file.close()
    return
```

```python
def generate_Udt_SCHW_circuit(ionorder,output_cal_list=False):
    L=[str(i) for i in ionorder];
    nqubit=6;
    statei=Quantum_State(nqubit)
    Set_Mapping(ionorder)
    mu = 0.1;
    x = 0.6;
    dt = 1;
    N = 10;
    
    cal_list=[]
    if(output_cal_list):
        cal_name="3-site-cal"+''.join(L)+".txt"
        write_cal=open(circuit_path+cal_name,'w')

    U1=Quantum_Circuit(nqubit,"1-evol")

    ## odd-leading terms
    for i in range(0,nqubit-1,2):
    #     print(i)
        U1.Add_Gate(Quantum_Gate("XA",i,i+1,angle=0.5*x*dt/pi))  
        U1.Add_Gate(Quantum_Gate("RZ",i,angle=2))
        U1.Add_Gate(Quantum_Gate("RZ",i+1,angle=2))
        U1.Add_Gate(Quantum_Gate("XA",i,i+1,angle=0.5*x*dt/pi))
        U1.Add_Gate(Quantum_Gate("RZ",i,angle=-2))
        U1.Add_Gate(Quantum_Gate("RZ",i+1,angle=-2))
        if(output_cal_list):
            M1=Quantum_Circuit(nqubit,"3-spatial")
            M1.Add_Gate(Quantum_Gate("XX",i,i+1))
            cal_seq = M1.GatesLab_Sequence()
            cal_seq_str = str(cal_seq)
            if (cal_seq_str[2]>cal_seq_str[3]):
                cal_seq_str = cal_seq_str[0:2]+cal_seq_str[3]+cal_seq_str[2]+cal_seq_str[4:]
            if cal_seq_str not in cal_list:
                cal_list.append(cal_seq_str)
                write_cal.write(cal_seq_str+"\n")

    #
    for i in range(1,nqubit-1,2):
    #     print(i)
        U1.Add_Gate(Quantum_Gate("XA",i,i+1,angle=0.5*x*dt/pi))  
        U1.Add_Gate(Quantum_Gate("RZ",i,angle=2))
        U1.Add_Gate(Quantum_Gate("RZ",i+1,angle=2))
        U1.Add_Gate(Quantum_Gate("XA",i,i+1,angle=0.5*x*dt/pi))
        U1.Add_Gate(Quantum_Gate("RZ",i,angle=-2))
        U1.Add_Gate(Quantum_Gate("RZ",i+1,angle=-2))
        if(output_cal_list):
            M1=Quantum_Circuit(nqubit,"3-spatial")
            M1.Add_Gate(Quantum_Gate("XX",i,i+1))
            cal_seq = M1.GatesLab_Sequence()
            cal_seq_str = str(cal_seq)
            if (cal_seq_str[2]>cal_seq_str[3]):
                cal_seq_str = cal_seq_str[0:2]+cal_seq_str[3]+cal_seq_str[2]+cal_seq_str[4:]
            if cal_seq_str not in cal_list:
                cal_list.append(cal_seq_str)
                write_cal.write(cal_seq_str+"\n")
    ## ZZ and Z terms
    for i in range(nqubit):
        U1.Add_Gate(Quantum_Gate("SK1Y",i,angle=-2))
    for i in range(1,nqubit-1):
        for j in range(i):
    #         print(j,i)
    #         print(nqubit-i-1)
            U1.Add_Gate(Quantum_Gate("XA",i,j,angle=0.5*(nqubit-i-1)*dt/pi))
            if(output_cal_list):
                M1=Quantum_Circuit(nqubit,"3-spatial")
                M1.Add_Gate(Quantum_Gate("XX",i,j))
                cal_seq = M1.GatesLab_Sequence()
                cal_seq_str = str(cal_seq)
                if (cal_seq_str[2]>cal_seq_str[3]):
                    cal_seq_str = cal_seq_str[0:2]+cal_seq_str[3]+cal_seq_str[2]+cal_seq_str[4:]
                if cal_seq_str not in cal_list:
                    cal_list.append(cal_seq_str)
                    write_cal.write(cal_seq_str+"\n")
    for i in range(nqubit):
        U1.Add_Gate(Quantum_Gate("SK1Y",i,angle=2))

    U1.Add_Gate(Quantum_Gate("AZ",0,angle=-(mu+3)*dt/pi)) 
    U1.Add_Gate(Quantum_Gate("AZ",1,angle=(mu-2)*dt/pi))
    U1.Add_Gate(Quantum_Gate("AZ",2,angle=-(mu+2)*dt/pi)) 
    U1.Add_Gate(Quantum_Gate("AZ",3,angle=(mu-1)*dt/pi))
    U1.Add_Gate(Quantum_Gate("AZ",4,angle=-(mu+1)*dt/pi))
    U1.Add_Gate(Quantum_Gate("AZ",5,angle=mu*dt/pi))
    seq = U1.GatesLab_Sequence()
#     print(str(seq))
    if(output_cal_list):
        write_cal.close()
    return str(seq)
```

```python
def generate_SCHW3_explicit_circuit(ionorder,dt=1):
    L=[str(i) for i in ionorder];
    nqubit=6;
    statei=Quantum_State(nqubit)
    Set_Mapping(ionorder)
    Nstep=7

    sim_file_name="donotrun_010101_dt="+str(dt)+"_eo"+''.join(L)+".txt"
    write_sim_file=open(circuit_path+sim_file_name,'w')
    
    B1=Quantum_Circuit(nqubit,"SCHW3block")
    B1.Add_Gate(Quantum_Gate("SK1X",1,angle=1))
    B1.Add_Gate(Quantum_Gate("SK1X",3,angle=1))
    B1.Add_Gate(Quantum_Gate("SK1X",5,angle=1))
    seq = B1.GatesLab_Sequence()
    seq_str=str(seq)
    write_sim_file.write(seq_str+"\n")
    Udt_str=generate_Udt_SCHW_circuit(ionorder,True)
    for i in range(Nstep):
        seq_str+=Udt_str
        write_sim_file.write(seq_str+"\n")
    write_sim_file.close()
    return
```
